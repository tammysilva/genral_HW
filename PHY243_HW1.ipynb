##Production of Fibonacci Sequence

**Part 1: Rewrite the function with for loop**

```python
def Fib(n):
  result1, result2 = 1, 1
  for i in range(n):
      result1, result2 = result2, result1 + result2
  return result1
Fib(9)
```

**Part 2: Which function is actually faster? (Explain without running the codes)**
*   For loop is faster.
*   According to the codes, the time complexity of the for loop one is O(n) whilst the time complexity of the recursive one is from O(n) to O(2^n). The bigger the n is, the differences between the two runtimes will be more obvious.



```python
def Fib_rec(n=0):
  if n==0:
    return 1
  elif n==1:
    return 1
  else:
    return Fib_rec(n-1)+Fib_rec(n-2)

Fib_rec(9)
```

**Part 3: Write a code to time the average time for k times function call. A function
that takes three arguments (function to time (Fib rec), input of the function (n), number of runs(k)) and run the Fib rec function k times for the input of n and returns the
average time.**

```python
import time
```

```python
def timer(n, k, f=Fib):
  x=time.time()
  for i in range(k):
    f(n)
  average_time = (time.time() - x)/k
  return average_time

print(f'The runtime for iterative fibonacci is {timer(5,10,Fib)}')
```

```python
print(f'The runtime for recursive fibonacci is {timer(5,10,Fib_rec)}')
```

**Part 4: Make a plot in which the x-axis is the value of the input function n and the
y-axis is the average time (output of the previous function), for both recursive and nonrecursive Fibonacci. (Both in the same plot; also use matplotlib package for making
the plots)**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = np.arange(10,100)
```

```python
plt.figure(figsize=(15,15))
n = np.arange(0,20)
t_fib = [timer(i, k=10, f=Fib) for i in n]
print(f'Average time for iterative fibonacci are {t_fib}')
t_fib_rec = [timer(i, k=10, f=Fib_rec) for i in n]
print(f'Average time for recursive fibonacci are {t_fib_rec}')
plt.plot(n, t_fib)
plt.plot(n, t_fib_rec)
plt.title("Average Runtime for Iterative Fibonacci and Recursive Fibonacci", fontsize = 20)
plt.legend(['Iterative Fibonacci', 'Recursive Fibonacci'])
plt.xlabel("The Input of n", fontsize = 16)
plt.ylabel("Average Time", fontsize = 16)
plt.show()
```
